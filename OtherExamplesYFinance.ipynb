```python
#The swiss knife for data science
import pandas as pd 
#For mathematical calculations
import numpy as np
import matplotlib.pyplot as plt # draw charts
import seaborn as sns #draw charts like a corr map
sns.set_style('whitegrid')
%matplotlib inline
import datetime
from scipy import stats #optimzing problems
import math
import yfinance as yf
```

```python
start = datetime.datetime(2015, 1, 1)
end = datetime.datetime(2025, 4, 10)
```

```python
#dax = yf.download('^GDAXI',rounding ="True", start=start, end=end, auto_adjust=False)
#bovespa = yf.download('^BVSP',rounding ="True", start=start, end=end, auto_adjust=False)
#sp500 = yf.download('^GSPC',rounding ="True", start=start, end=end, auto_adjust=False,interval='1mo')
```

```python
indices = ['^GDAXI', '^BVSP', '^GSPC']
dataindices = yf.download(indices,rounding ="True", start=start, end=end, auto_adjust=False,interval='1mo',group_by='indices')

# extract the closing prices
adjclose_prices = pd.DataFrame()
for index in indices:
    adjclose_prices[index] = dataindices[index]['Adj Close']
```

```python
pct_adjclose_prices = adjclose_prices.pct_change()

# Display the resulting DataFrame
print(pct_adjclose_prices)
```

```python
# compute the correlation matrix
corr_indeces = pct_adjclose_prices.corr()

# create a heatmap of the correlation matrix using Seaborn
sns.set(style='white')
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corr_indeces, annot=True, cmap='RdYlGn', vmin=-1, vmax=1, ax=ax)

# set the axis labels and title
ax.set_xlabel('Indices')
ax.set_ylabel('Indices')
ax.set_title('Correlation Matrix of Indices')

# display the plot
plt.show()
```
